```python
import numpy as np
import pandas as pd
#!pip3 install folium==0.5.0
import folium
from folium import plugins
print('Folium installed and imported!')
```

```python
path = '/content/drive/MyDrive/Data science/Mine/Maps-Vietnam_University_Locations /Vietnam university list.xlsx'
data = pd.read_excel(path)
data.head()
```

```python
# Viet Nam latitude and longitude values
latitude = 14.058324
longitude = 108.277199
```

```python
#Map 1: Pop-up text to each marker on the map
# instantiate a feature group for the locations in the dataframe
locations = folium.map.FeatureGroup()
#Create map and display it
vietnam_map_1 = folium.Map(location=[latitude, longitude], zoom_start=5) #tiles='Stamen Terrain #zoom_start = zoom level
#tiles:'Stamen Toner' => high-contrast B+W (black and white); 'Stamen Terrain' => hill shading and natural vegetation colors
#Loop through all locations and add each to the locations feature group
for lat, lng, in zip(data.LATITUDE, data.LONGITUDE):
    locations.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=8, #how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.8
        )
    )
#Add pop-up text to each marker on the map
latitudes = list(data.LATITUDE)
longitudes = list(data.LONGITUDE)
labels = list(data.CLASS)
for lat, lng, label in zip(latitudes, longitudes, labels): 
  folium.Marker([lat, lng], popup=label).add_to(vietnam_map_1)    
# add locations to map
vietnam_map_1.add_child(locations)
```

```python
#Map 2: Pop-up text to each marker and group the markers into different clusters on the map
#Create map and display it
vietnam_map = folium.Map(location=[latitude, longitude], zoom_start=5) 
#Instantiate a mark cluster object for the locations in the dataframe
locations = plugins.MarkerCluster().add_to(vietnam_map)
#Loop through the dataframe and add each data point to the mark cluster
for lat, lng, label, in zip(data.LATITUDE, data.LONGITUDE, data.CLASS):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(locations)
# display map
vietnam_map
```
