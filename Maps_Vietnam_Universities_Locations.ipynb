```python
import numpy as np
import pandas as pd
#!pip3 install folium==0.5.0
import folium
from folium import plugins
print('Folium installed and imported!')
```

```python
path = '/content/drive/MyDrive/Data science/Mine/Maps-Vietnam_University_Locations /Vietnam university list.xlsx'
data = pd.read_excel(path)
data.head()
```

```python
#Pop-up text to each marker and group the markers into different clusters on the map
# Viet Nam latitude and longitude values
latitude = 14.058324
longitude = 108.277199
#Create map and display it
vietnam_map = folium.Map(location=[latitude, longitude], zoom_start=5) #tiles='Stamen Terrain #zoom_start = zoom level
#tiles:'Stamen Toner' => high-contrast B+W (black and white); 'Stamen Terrain' => hill shading and natural vegetation colors

#Instantiate a mark cluster object for the locations in the dataframe
locations = plugins.MarkerCluster().add_to(vietnam_map)
#Loop through the dataframe and add each data point to the mark cluster
for lat, lng, label, in zip(data.LATITUDE, data.LONGITUDE, data['UNIVERSITY NAME']):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(locations)
# display map
vietnam_map
```
