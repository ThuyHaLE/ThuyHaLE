```python
#Import some library
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import urllib
import urllib.request
from PIL import Image

from google.colab import drive
drive.mount ('/content/drive')
from google.colab import files
```

```python
#Save mask to belle_mask
belle_mask = np.array(Image.open(urllib.request.urlopen(
    'https://drive.google.com/uc?export=download&id=1Gj5JPTRHn2Oj3fDTU6ZpPbCgzVYkAu9a')))
fig = plt.figure(figsize=(15, 15))
plt.imshow(belle_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#Open the text file and read it into a variable belle_novel
belle_novel = urllib.request.urlopen(
    'https://drive.google.com/uc?export=download&id=1qZiBQkdpbUwJ1vmVuibp6jNkhTtmL0Nk').read().decode("utf-8")

#Let's use the stopwords that we imported from word_cloud. 
#We use the function set to remove any redundant stopwords.
stopwords = set(STOPWORDS)

#Instantiate a word cloud object
belle_wc = WordCloud(
    background_color='black',
    max_words=2000000,
    mask=belle_mask, 
    colormap='cool',
    font_path='/content/drive/MyDrive/Data science/Mine/Word Clouds-Belle/ ProN W4.ttc',
    stopwords=stopwords).generate(belle_novel) # generate the word cloud

#Display and save the word cloud
fig = plt.figure(figsize=(20, 20))
plt.imshow(belle_wc, interpolation='bilinear')
plt.axis('off')
plt.savefig("Belle the dragon and the freckled princess.png")
plt.show()
files.download('Belle the dragon and the freckled princess.png')
```
